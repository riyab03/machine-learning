CHANGING CATEGORICAL VALUES TO NUMERICAL 

```python
label_encoder = LabelEncoder()
categorical_columns = kidney_data.select_dtypes(include=['object']).columns
kidney_data[categorical_columns] = kidney_data[categorical_columns].apply(lambda col: label_encoder.fit_transform(col.astype(str)))
```

```python
kidney_data.drop('id', axis=1, inplace=True)
kidney_data.drop('appet', axis=1, inplace=True)
```

```python
kidney_data['classification'].value_counts()
```

```python
X = kidney_data.drop(columns = 'classification',axis=1)
Y = kidney_data['classification']
print(Y)
```

FEATURES WITH LEAST IMPORTANCE USING VARIANCE THRESHOLD AND RANDOM FOREST

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
threshold = 0.01
variance_selector = VarianceThreshold(threshold=threshold)

# Fit the selector to the data
variance_selector.fit(X)

# Get the indices of non-constant features
non_constant_indices = variance_selector.get_support(indices=True)

# Get the names of constant features
constant_features = set(X.columns) - set(X.columns[non_constant_indices])

# Print the constant features
print("Constant or Near-Constant Features:", constant_features)
```

```python
kidney_data.drop('sg', axis=1, inplace=True)
```

```python
correlated_features = set()
correlation_matrix = kidney_data.corr()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
```

```python
print("Highly Correlated Features:", correlated_features)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X, Y)

feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances = feature_importances.sort_values(ascending=False)

print(feature_importances)
```

```python
kidney_data.drop(['pcc','pot','cad','wc','ba','age','dm','pe','bp','su','ane'],axis=1,inplace=True)
```

```python
print(kidney_data.columns)
```

```python
X = kidney_data.drop(columns = 'classification',axis=1)
Y = kidney_data['classification']
print(Y)
```

TRANING AND TESTING

```python

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

TRAINING DATASET PREDICTION ACCURACY 

```python
from sklearn.metrics import accuracy_score
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print(training_data_accuracy)
```

```python
X_test_prediction= model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print(testing_data_accuracy)
```
